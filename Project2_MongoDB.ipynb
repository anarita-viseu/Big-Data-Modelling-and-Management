##### 2.1. "maximum_nights" field

```python
# Counting the number of times each unique value of maximum_nights appears
value_counts = db.listing_customer.aggregate([
    # Converting the maximum_nights field to integer using toInt()
    {"$project": {"maximum_nights_int": {"$toInt": "$maximum_nights"}}},
    # Grouping by the field maximum_nights_int and counting the number of times each value appears in the collection
    {"$group": {"_id": "$maximum_nights_int", "count": {"$sum": 1}}},
    # Sorting the results by descending order of the "maximum_nights_int" field
    {"$sort": {"_id": -1}},
    {"$project": {"_id":0, "Maximum Nights": "$_id", "Count": "$count"}}
])


# Printing the output using pprint
pprint(list(value_counts))
```

Our team noticed some values that are strange, for instance, ``Maximum Nights`` higher than 3 years, or the values "333333", "2147483647" or "1234567890".

Additionally, we also noticed that the most common ``Maximum Nights`` is 1125, with 3137 Airbnbs with that number in the maximum_nights field. Because of that, our team decided to delete every record with ``Maximum Nights`` higher than 1125. 

To perform this, we will take into consideration that the ``Maximum Nights`` field is stored as a string, therefore we will use the ``parseInt`` funtion that converts a string to an integer. Then, we remove all fields with a higher value than 1125.

```python
result = db.listing_customer.find({
    # Convert maximum_nights to an integer, and filter the documents where Maximum Nights is greater than 1125
    "$where": "parseInt(this.maximum_nights) > 1125"
}, {"_id": 0, "maximum_nights": 1})

# Using the delete_many function, we delete the documents where Maximum Nights is greater than 1125
result = db.listing_customer.delete_many({
    "$where": "parseInt(this.maximum_nights) > 1125"
})

# Printing the number of documents that were deleted
print(f"{result.deleted_count} documents deleted.")
```

Now, we will check for duplicates in the ``summary`` field, since it is strange if two different Airbnbs to have the same summary.

```python
# Check for Duplicates in the field summary
duplicates_summary = db.listing_customer.aggregate([
    # Grouping by the "summary" field and counting how many times each id appears in the collection listing_customer
    {"$group": {"_id": "$summary", "count": {"$sum": 1}}},
    # Show only if there are duplicates (greater than 1)
    {"$match": {"count": {"$gt": 1}}},
    # Sun the number of duplicates
    {"$group": {"_id": None, "Nr of duplicates": {"$sum": 1}}},
    # Project the "Nr of duplicates" field
    {"$project": {"_id": 0, "Nr of duplicates": 1}}
])

# Check the count 
results = list(duplicates_summary)
print(results)
```

There are 34 duplicates in the ``summary`` field. However, it might not be a problem if the duplicates belong to the same host, who might have similar or even exact the same listings. For example, an apartment building where the host is listing the different appartments he has, or terraced houses that are all lookalike. Therefore, we decided not to delete these duplicates.

We will check for duplicated ``summary`` fields between different hosts, because this might not be an error, and therefore should be deleted.

```python
duplicates_summary = db.listing_customer.aggregate([
    # Group by host_id and summary, and count the number of occurrences
    {"$group": {"_id": {"host_id": "$host_id", "summary": "$summary"}, "count": {"$sum": 1}}},
    # Group by summary only, and count the number of groups with more than one host_id
    {"$group": {"_id": "$_id.summary", "Nr of duplicates": {"$sum": {"$cond": [{"$gt": ["$count", 1]}, 1, 0]}}}},
    # Show only if there are duplicates (greater than 1)
    {"$match": {"Nr of duplicates": {"$gt": 1}}},
    # Project the number of duplicates
    {"$project": {"_id": 0, "Nr of duplicates": 1}}
])

# Check the count 
results = list(duplicates_summary)
pprint(results)
```

There are 6 duplicated `summary` fileds among different hosts. Considering that this might be an error, we decided to delete these 6 documents.

```python
# Find the duplicated documents
duplicates = db.listing_customer.aggregate([
    # Group by host_id and summary, and count the number of occurrences
    {"$group": {
        "_id": {"host_id": "$host_id", "summary": "$summary"},
        "ids": {"$addToSet": "$_id"},
        "count": {"$sum": 1}
    }},
    # Show only if there are duplicates (greater than 1)
    {"$match": {"count": {"$gt": 1}}}
])

# Delete the duplicated documents
for doc in duplicates:
    ids_to_delete = doc['ids'][1:]
    db.listing_customer.delete_many({"_id": {"$in": ids_to_delete}})
```

This collection lists all unique amenities currently listed in properties (across all documents) ordered by most common to least common.

```python
# 2. Creating a function that shows the number of reviews for a given reviewer ID or Name

def review_count_by_identity(reviewer_identity):
    """ This function returns the number of reviews for a given reviewer ID or Name. """
    # if the input is an ID
    if str(reviewer_identity).isdigit():
        # Count the number of documents in the collection 'listing_reviews' where the field 'reviewer_id' matches the given ID
        # Using str() in case the user gives an integer instead of a string
        review_count = db.listing_reviews.count_documents({"reviewer_id": str(reviewer_identity)})
        # Print the result
        print('The reviewer with ID '+str(reviewer_identity)+' wrote '+str(review_count)+' reviews.')
        
    # if the input is the name of a reviewer
    else:
        # Create a list with each reviewer with the given name and the respective number of reviews
        reviewers= list(db.listing_reviews.aggregate([
            # Get the documents where the field 'reviewer_name' matches the given name
            {"$match": {"reviewer_name": reviewer_identity}},
            # Group by the field 'reviewer_id' and count the number of reviews
            {"$group": {"_id": "$reviewer_id", "Count": {"$sum": 1}}},
            # Sort by the number of reviews in descending order
            {"$sort": {"Count": -1}},
            # Specify the fields '_id' and 'Count'
            {"$project": {"_id": 1, "Count": 1}}]))
        for i in reviewers:
            print('The reviewer with name '+str(reviewer_identity)+' and ID '+str(i['_id'])+' wrote '+str(i['Count'])+' reviews.')
```

### Question 11
We wish to be able to have better access to information about transaction, we wish to develop a search engine that can calculate the average value of transactions in a given period of time quickly for a given property.

```python
# Define the function that calculates the average value of transactions for the specific period of time and property given
def avg_transaction_value(property_id, start_date, end_date):
    """ This function calculates the average value of transactions of a given property and for a given period of time."""
    
    # Define the aggregation pipeline to retrieve the average transaction value
    pipeline = [
        # Unwind the transactions array
        {"$unwind": "$transactions.transactions"},
        # Project the necessary fields and convert the date to a string
        {"$project": {
            "_id": 1,
            "transactions.transactions.date": {"$dateToString": {"format": "%Y, %m, %d, %H, %M", "date": "$transactions.transactions.date"}},
            "transactions.transactions.price": {"$toDouble": "$transactions.transactions.price"}
        }},
        # Filter by the specified property and the specified time period
        {"$match": {"$and": [{"_id": property_id},
                             {"transactions.transactions.date": {"$gte": start_date.strftime("%Y, %m, %d, %H, %M")}},
                             {"transactions.transactions.date": {"$lte": end_date.strftime("%Y, %m, %d, %H, %M")}}]}},
        # Use "_id": None to calculate the average price across all documents
        #(the id is always the same because it was defined at the beginning)
        {"$group": {"_id": None, "avg_price": {"$avg": "$transactions.transactions.price"}}}
    ]
    
    # Execute the aggregation pipeline and store the results in the 'result' variable
    result = list(db.listingsAndReviews_new.aggregate(pipeline))
    
    # If the result set is empty, print an error message
    if len(result) == 0:
        print("There are no documents in the collection that match the query.")
    # If the result set is not empty, print the average transaction price for the given property and time period
    else:
        print(f"The average transaction price for property {property_id} between {start_date.strftime('%Y-%m-%d')} and {end_date.strftime('%Y-%m-%d')} is {result[0]['avg_price']:.2f}.")
    
```

```python
# Top 10 cities and their top 3 properties (based on their average rating)
top10_cities = list(db.listingsAndReviews_new.aggregate([
    # First we want to group by cities so that we can find the top 10 using the sort
  { '$group': { '_id': "$address.market", 
               'count': { '$sum': 1 },
              } },
  { '$sort': { 'count': -1 } },
  { '$limit': 10 },
    
    # We use lookup to make a list of all the properties in each city so that we can order them using review ratings
  { '$lookup': { 'from': "listingsAndReviews_new", 
                'localField': "_id", 
                'foreignField': "address.market", 
                'as': "properties" } },
    # We use unwind to have a different document for each property
  { '$unwind': "$properties" },
    # Again so that we can have all the different reviews separated 
  { '$unwind': "$properties.reviews" },
  { '$group': { '_id': "$properties._id", 
               'avg_rating': { '$max': "$properties.review_scores_rating" } } },
    # Sorting the properties by highest review scores rating
  { '$sort': { 'avg_rating': -1 } },
  { '$group': { '_id': "$_id", 
               'avg_rating': { '$first': "$avg_rating" } } },
    # Then in each property we also want a list of informations on that specific property
  { '$lookup': { 'from': "listingsAndReviews_new", 
                'localField': "_id", 
                'foreignField': "_id", 
                'as': "property_info" } },
    # We chose some variables that we thought were more useful for the summary webpage
  { '$project': { 'city': "$property_info.address.market", 
                 'property_id': "$_id",
                 'avg_rating': 1, #The rating of the property
                 'number_of_reviews': '$property_info.number_of_reviews', #Number of reviews 
                 'price':'$property_info.price', #Price
                 # If the host is a superhost, meaning he will probably be trustworthy
                 'host_is_superhost':'$property_info.host_is_superhost',
                 'property_type':'$property_info.property_type', # The type of property
                 'property_name': "$property_info.name" } }, # Property name
    # Grouping the documents by city
  { '$group': { '_id': "$city", 
               'properties': { '$push':  # Creating a new field called properties
                              { 
                                   'Property ID': "$property_id", 
                                   'Name': "$property_name", 
                                   'Rating': "$avg_rating",
                                   'Number of reviews':'$number_of_reviews',
                                   'Superhost':'$host_is_superhost',
                                   'Type of property':'$property_type',
                                   'Price': '$price'} } } },
  { '$project': { '_id': 0, 
                 'City': "$_id", # Making the id the City
                 'Properties': { '$slice': [ "$properties", 3 ] } } } # Top 3 properties
]))

pprint(top10_cities)
```
